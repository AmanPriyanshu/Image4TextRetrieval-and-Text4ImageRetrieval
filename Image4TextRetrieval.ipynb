# Loading the model

```python
from transformers import AutoTokenizer, CLIPModel, AutoProcessor

model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
tokenizer = AutoTokenizer.from_pretrained("openai/clip-vit-base-patch32")
processor = AutoProcessor.from_pretrained("openai/clip-vit-base-patch32")
```

## Download Cat Image:

```python
import requests

url = 'https://raw.githubusercontent.com/AmanPriyanshu/Image4TextRetrieval-and-Text4ImageRetrieval/main/cat.jpg'
response = requests.get(url)
with open('cat.jpg', 'wb') as file:
    file.write(response.content)
```

## Trying out retrieval

### Encoding all texts

```python
text_description_list = ["a photo of a cat", "a photo of a dog"]

inputs = tokenizer(text_description_list, padding=True, return_tensors="pt")
text_features = model.get_text_features(**inputs)
```

```python
text_features.shape
```

### Encoding image

```python
from PIL import Image

image = Image.open("cat.jpg")
```

```python
inputs = processor(images=image, return_tensors="pt")
image_features = model.get_image_features(**inputs)
```

```python
image_features.shape
```

## Retrieve closest text

```python
import torch
import torch.nn.functional as F

cos_sim = F.cosine_similarity(text_features, image_features, dim=1)

# Find the index of the highest similarity
closest_index = torch.argmax(cos_sim).item()

print("Cosine similarities:", cos_sim)
print("Index of the closest tensor:", closest_index)
print("Retrieved:", text_description_list[closest_index])
```
